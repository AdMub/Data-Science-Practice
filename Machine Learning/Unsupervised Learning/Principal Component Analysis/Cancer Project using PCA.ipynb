```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer["data"], columns=cancer["feature_names"])
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

# PCA

```python
from sklearn.decomposition import PCA # Decomposing into Principal Component
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pa = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pa.shape
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_pa[:,0], x_pa[:,1])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_pa[:,0], x_pa[:,1])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_pa[:,0], x_pa[:,1], c=cancer["target"])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_pa[:,0], x_pa[:,1], c=cancer["target"], cmap="plasma")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer["feature_names"])
df_comp
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df_comp, cmap="plasma")
```
